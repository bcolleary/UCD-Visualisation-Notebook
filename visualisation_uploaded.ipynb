# Visualisation

Note: This will use datasets from DataCamp (Summer Olympics 2016 and Austin Weather)

## Importing Modules

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Scatter Plot

```python
summer2016 = pd.read_csv('summer2016.csv')
summer2016.head()
```

Continuous variables

## Data Types

To know **Categorical** or **Numerical**

```python
summer2016.info()
```

### Matplotlib

```python
plt.scatter(summer2016['Height'],summer2016['Weight'], s=2); #change size of bar
```

### Let's add labels

```python
plt.scatter(summer2016['Height'],summer2016['Weight'], s=2)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Olympic Athletes");
#change size of bar
```

# Adding Vertical Lines and limiting the axes

```python
plt.subplots(figsize=(10,5)) #change the size of the figure
plt.scatter(summer2016["Height"], summer2016["Weight"], s=2)


plt.xlim(left=0,right=1000) #To specify range of axis to be shown (optional parameters)
#You can choose to ignore any of the parameters above (e.g left or right) and it'll automatically use the default limit


plt.axvline(x=100, color="green", label="vert") #to include a vertical line (e.g targets)
plt.axhline(y=100, color="red", label="horz")  #to include a horizontal line (e.g. targets)


plt.xticks(rotation=45) #rotate the values shown on the x-axis
plt.yticks(rotation=45) #rotate the values shown on the y-axis

plt.xlabel("Height (cm)") #label x-axis
plt.ylabel("Weight (kg)") #label y-axis
plt.title("Olympic Athletes")

plt.legend() #show labels 
plt.show()
```

### Seaborn

```python
sns.scatterplot(data=summer2016, x='Height',y='Weight', hue='Sex');

#add another column
```

## Line Plot

Let's try a different dataset- Austin weather dataset

```python
austin = pd.read_csv('austin_weather.csv')
austin.head()
```

### Matplotlib

```python
plt.plot(austin['DATE'],austin['MLY-CLDD-BASE45'])

#makes sense --> it has a some form of order --> see PPT
#Ordinal vs Continuous
```

```python
#plt.subplots(figsize=(10,5))

plt.plot(austin['DATE'],austin['MLY-CLDD-BASE45'])
plt.plot(austin['DATE'],austin['MLY-CLDD-BASE70'])

plt.xlabel('Date (months)')
plt.ylabel('Amount of Rain (mm)')
plt.title('Rainfall across Year')
plt.show()
```

## Bar Chart

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
         'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [1100, 1200, 1300, 1350, 1400, 1387,
         1499, 1574, 1611, 1644, 1587, 1721]
```

```python
plt.bar(month, sales)
plt.show()
```

**Horizontal bar chart**

```python
month = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
         'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [1100, 1200, 1300, 1350, 1400, 1387,
         1499, 1574, 1611, 1644, 1587, 1721]
plt.barh(month, sales)
plt.show()
```

```python
#number = "123456789"
```

```python
#number[::-1]
```

```python
plt.barh(month[::-1], sales[::-1])
plt.show()
```

### Seaborn

```python
sns.barplot(x=month, y=sales);
```

**Horizontal**

```python
sns.barplot(y=month, x=sales);
```

Horizontal - change colour

```python
sns.barplot(y=month, x=sales, color="blue");
```

## Count Plot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

### Categorical values

### Count of Male and Female on Titanic

```python
titanic["sex"].value_counts()
#categorical
```

```python
titanic["sex"].value_counts().index
```

```python
titanic["sex"].value_counts().values
```

## Pairplot

Runs a scatter plot against all your numerical data 

Looks at your dataset, sees how many numerical columns you have and plots them against each other

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```
